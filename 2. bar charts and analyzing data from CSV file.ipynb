```python
with open('dataforhist.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

print(language_counter.most_common(15))
'''here each item of the list will be containing the tuple of two items, 
one is language and other one is the count'''
```

>*  we want to  convert the above list which contain tuples into two lists, one containg the list of laguages and another containg the list of counts of each languages 

```python
with open('dataforhist.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
print(languages)
print(popularity)
```

>* now let's plot this data

```python
plt.bar(languages, popularity)

plt.title('Most Popular Programming Languages')
plt.xlabel('Programming Languages')
plt.ylabel('Number of People Who Use')

plt.show()
```

>*  as we can see if we have to many items then it's not  convinient to use the vertical bar chart because item names are overlapping each othe so instead the we will use the horizontal bar chart
>* use - plt.hbar()

```python
plt.barh(languages, popularity)

plt.title('Most Popular Programming Languages')
plt.xlabel('Number of People Who Use')
plt.tight_layout() # to concise the size of the chart

plt.show()
```

>* to get the most popular language first we will reverse our list 

```python
with open('dataforhist.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
languages.reverse() # it will reverse the list inplace (permanently)
popularity.reverse()

print(languages)
print(popularity)
```

```python
plt.barh(languages, popularity)

plt.title('Most Popular Programming Languages')
plt.xlabel('Number of People Who Use')

plt.show()
```

```python
import numpy as np
import pandas as pd 
from collections import Counter
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dataforhist.csv')
```

```python
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses :
    language_counter.update(response.split(';'))
    

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
languages.reverse() # it will reverse the list inplace (permanently)
popularity.reverse()

print(languages)
print(popularity)
```

```python
plt.barh(languages, popularity)

plt.title('Most Popular Programming Languages')
plt.xlabel('Number of People Who Use')

plt.show()
```
